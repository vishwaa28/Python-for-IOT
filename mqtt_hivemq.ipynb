```python
#publisher1:
import paho.mqtt.client as mqtt
from random import uniform
import time

mqtt_broker = "broker.hivemq.com"
client = mqtt.Client("temperature-inside")
client.connect(mqtt_broker)

while True:
    random_temp = uniform(20.0, 22.0)
    client.publish("temperature/inside", random_temp)
    print(f"Just published {random_temp} to topic temperature/inside")
    time.sleep(1)
```

```python
#publisher2:
import paho.mqtt.client as mqtt
from random import randrange
import time

mqtt_broker = "broker.hivemq.com"
client = mqtt.Client("temperature-outside")
client.connect(mqtt_broker)

while True:
    random_temp = randrange(12)
    client.publish("temperature/outside", random_temp)
    print(f"Just published {random_temp} to topic temperature/outside")
    time.sleep(1)
```

```python
#subscriber
import paho.mqtt.client as mqtt
import time

def on_message(client, userdata, message):
    print("Received message:", str(message.payload.decode("utf-8")))

mqtt_broker = "broker.hivemq.com"
client = mqtt.Client("smartphone")
client.connect(mqtt_broker)

client.loop_start()
client.subscribe("temperature/#")
client.on_message = on_message

# Keep the script running
try:
    while True:
        time.sleep(1)
except KeyboardInterrupt:
    client.loop_stop()
    client.disconnect()
```
